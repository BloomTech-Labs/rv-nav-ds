```
first_filter= [
    'LAT_016','LONG_017','STATE_CODE_001',
    'TOLL_020','MIN_VERT_CLR_010', 'VERT_CLR_OVER_MT_053',
    'VERT_CLR_UND_REF_054A','VERT_CLR_UND_054B',
    'TRAFFIC_LANES_UND_028B'
]
```

```
def filter_low(df, min_clr, max_clr):
  'filters df by overpass bridges and clearance threshold'
  df = df[df['TRAFFIC_LANES_UND_028B'] != 0]
  df = df[df['MIN_VERT_CLR_010'] < min_clr]
  df = df[df['MIN_VERT_CLR_010'] > max_clr]
  df = df[first_filter]
  return df
```
